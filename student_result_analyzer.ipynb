```python
import numpy as np

print("="*50)
print("STUDENT MARKS ANALYSER")
print("="*50)

dataset = np.random.randint(25,101,size=(20,3))
print(dataset)

print("Average marks per student :",np.round(dataset.mean(axis=1),2))
print("Average marks per subject :",np.round(dataset.mean(axis=0),2))
print("highest score : ",dataset.max())
print("lowest score : ",dataset.min())
passed = dataset[np.any(dataset>=40,axis=1)]
print("passed :",passed)
top_student = dataset[np.any(dataset>80,axis=1)]
print("top students are :",top_student)

```
